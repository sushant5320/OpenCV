```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('DATA/sammy_face.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
edge = cv2.Canny(img,threshold1 = 150, threshold2 = 150)
```

```python
plt.imshow(edge)
```

```python
blurred_img = cv2.blur(img, ksize = (5,5))
```

```python
edge = cv2.Canny(blurred_img,threshold1 = 100, threshold2 = 100)
```

```python
plt.imshow(edge)
```

```python
lower = int(max(0,0.7*np.median(img)))
upper = int(min(255,1.3*np.median(img)))
```

```python
edge = cv2.Canny(blurred_img,threshold1 = lower+20, threshold2 = upper+45)
plt.imshow(edge)
```
