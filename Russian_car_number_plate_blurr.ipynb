```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('DATA/car_plate.jpg',0)
img_copy = img.copy()
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
car_plate_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_russian_plate_number.xml')
```

```python
lst = []
def detect_plate(img):
    img_copy = img.copy()
    rect = car_plate_cascade.detectMultiScale(img_copy, scaleFactor = 1.2, minNeighbors = 5)

    
    for (x, y, w, h) in rect:
        cv2.rectangle(img_copy, (x,y), (x+w, y+h), (255, 255, 255), 1)
        lst.append(x)
        lst.append(y)
        lst.append(w)
        lst.append(h)
        

    return img_copy
```

```python
result = detect_plate(img)
```

```python
plt.imshow(result, cmap = 'gray')
```
